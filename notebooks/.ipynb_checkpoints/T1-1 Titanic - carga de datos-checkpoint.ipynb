```python
cols = data3.readline().strip().split(",")   #strip: elimina espacios en blanco, split: divide la linea de texto por un delimitador
n_cols = len(cols)
```

```python
n_cols
```

```python
counter = 0 
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
main_dict
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
print("El data set tiene %d filas y %d columnas" %(counter, n_cols))
```

### Ejemplo usandola libreria urllib3 para leer los datos desde una URL externa, perocesarlos y convertirlos a un dara frame de python antes de guardarlos en un CSV local

```python
# Primero importamos la libreria y hacemos la conexion con la web de los datos

import urllib3
http = urllib3.PoolManeger()
r = http.request('GET', url)
print("El estado de la respuesta es %d" %(r.status))
response = r.data

# El objeto response contiene un string binario, asi que lo convertimos a un string decodificado en UTF-8
str_data = response.decode('utf-8')

# Dividimos el string en un array de filas, separandolo por intros
lines = str_data.split("\n")

# La primera linea contiene la cabecera, asi que la extraemos
col_names = lines[0].split(",")
n_cols = len(col_names)

# Generamos un diccionario vacio donde irá la informacion procesada desde la URL externa

counter = 0
main_dict = {}
for col in col_names:
    main_dict[col] = []
    
# Procesamos fila a fila la informacion para ir rellenando el diccionario con los datos como lo hicimos antes

for line in lines:
    #Nos saltamos la primera fila que es la que contiene la cabecera y ya tenemos procesada
    if(counter > 0):
        #Dividimos cada string por las comas como elemento separador ( o segun corresponda)
        values = line.strip().split(",")
        #Añadimos cada valor a su respectiva columna del diccionario
        for i in range(len(col_names)):
            main_dict[col_names[i]].append(values[i])
    counter += 1
    
print("El data set tiene %d filas y %d columnas"%(counter, n_cols))

#Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
nombre_df = pd.DataFrame(main_dict)
print(nombre_df.head())

#Elegimos donde guardarlo
mainpath = "/Users/LenovoYoga/Documents/GitHub/Machine-Learning-Python/Machine-Learning-Python/datasets"
filename = "athletes/dowloaded_nombre."
fullpath = os.path.join(mainpath, filename)

#Lo guardamos en CSV, JSON, o EXCEL.
nombre_df.to_csv(fullpath+"csv")
nombre_df.to_json(fullpath+"json")
nombre_df.to_excel(fullpath+"xls")
```
